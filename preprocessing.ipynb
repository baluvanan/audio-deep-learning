```python
import librosa,  librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

Importing Audio 

```python
audio_input = "./audio/pre_preprocess_tut/blues.00000.wav"
```

```python
signal, sr = librosa.load(audio_input,sr=22050)
```

## Rendering waveform

```python
librosa.display.waveplot(signal, sr= sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

## Rendering Spectrum

Fast fourier transform will provide a Spectrum 

```python
fft = np.fft.fft(signal)
magnitude = np.abs(fft)
frequency = np.linspace(0,sr, len(magnitude))

plt.plot(frequency,magnitude)
plt.xlabel("frequency")
plt.ylabel("magnitude")
```

```python
print(len(magnitude))
print(len(fft))
```

## Rendering Spectrogram
Short time fourier transform will provide Spectrogram
n_fft = window of fft or the number of samples in a single fft 
hops_length = amount of move of the window

```python
n_fft = 2048
hop_length = 512  
```

```python
stft = librosa.core.stft(signal,n_fft=n_fft,hop_length=hop_length)
spectrogram = np.abs(stft)

librosa.display.specshow(spectrogram,sr=sr, hop_length=hop_length)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar()
plt.show
```

Most frequency do contribute to the loudness of the sound the way we percieve sound is logrithmic

```python
log_spectrogram = librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram,sr=sr, hop_length=hop_length)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar()
plt.show
```

# MFCC's

```python
MFCCs = librosa.feature.mfcc(signal,n_fft=n_fft,hop_length=hop_length,n_mfcc=13)

librosa.display.specshow(MFCCs,sr=sr, hop_length=hop_length)
plt.xlabel('Time')
plt.ylabel('MFCC')
plt.colorbar()
plt.show
```
